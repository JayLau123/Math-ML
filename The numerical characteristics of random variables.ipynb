## The numerical characteristics of random variables

### 随机变量的数字特征（芝大课程总结）

### Expectation of random variable 期望

A random variable $X$ is discrete, it has probability mass function $p_X(a)=P(X=a)$

The expectation of $X$ is

$$E(X)=\sum p_X(x)\ x$$

It is the probability-weighted average of values of $X$.

**Properties of expectation**

$E(aX+b)=aE(X)+b$

$E(X+Y)=E(X)+E(Y)$

$E(a_1X_1+a_2X_2+...+a_nX_n)=a_1E(X_1)+a_2E(X_2)+...+a_nE(X_n)$


### Variance of random variable 方差

Suppose that discrete random variable $X$ has expectation $E(X)$, its variance is 

$$D(X)=Var(X)=E[(X-E(X))^2]$$

More explicitly, this is given by

$$D(X)=\sum p_X(x)(x-E(X))^2$$

(P.S. 平方相比于绝对值更加好处理)

方差意义：衡量随机变量与其平均值的偏离程度

**It is the probability-weighted average deviation of $X$ from $E(X)$**

方差是随机变量$X$的函数$g(X)=(X-E(X))^2$的数学期望

$D(X)=\sum p_X(x)(x-E(X))^2$

$=E(X^2-2XE(X)+E(X)^2)=E(X^2)-2E(X)E(X)+E(X)^2=E(X^2)-E(X)^2$

If $X$ is a continuous random variable, so 

$$D(X)=\int_{-\infty}^{+\infty}(x-E(X))^2 f(x) dx$$



**Properties of variance**

$D(X)=0$ if and only if $P(X=E(X))=1$

$D(aX+b)=a^2D(X)$

$D(X+Y)=D(X)+D(Y)+2Cov(X,Y)$

when $X$ and $Y$ are uncorrelated, $Cov(X,Y)=0$

### Standard deviation of samples 标准差=均方差

#### 标准差=$\sqrt{方差}$

意义：因为方差与我们要处理的数据的量纲不一致，所以标准差能更好地描述数据与均值的偏离程度

#### A discrete random variable $X$


The values have different probabilities, let $x_1$ have probability $p_1$, $x_2$ have probability $p_2$, …, $x_N$ have probability $p_N$. In this case, the standard deviation will be
 
$$\sigma(X)=\sqrt{D(X)}=\sqrt{\sum_{i=1}^N p_i\left(x_i-\mu\right)^2}, \text { where } \mu=\sum_{i=1}^N p_i x_i$$

In the special case where X takes random values from a finite data set $x_1, x_2, …, x_N$, with each value having the **same probability**, the standard deviation is


$$\sigma(X)=\sqrt{\frac{1}{N}\left[\left(x_1-\mu\right)^2+\left(x_2-\mu\right)^2+\cdots+\left(x_N-\mu\right)^2\right]}, \text { where } \mu=\frac{1}{N}\left(x_1+\cdots+x_N\right)$$


$$\sigma(X)=\sqrt{\frac{1}{N} \sum_{i=1}^N\left(x_i-\mu\right)^2}, \text { where } \mu=\frac{1}{N} \sum_{i=1}^N x_i$$



#### A continuous random variable $X$

$$\mu = E(X)=\int_{-\infty}^{+\infty} x f(x) dx$$


The standard deviation of $X$ is


$$ \sigma(X) =\sqrt {D(X)}= \sqrt{E[(X-\mu)^2]}=\sqrt{\int_{-\infty}^{+\infty}(x-\mu)^2 f(x) dx}$$





**properties**

$\sigma(aX+b)=\sqrt{D(aX+b)}=\sqrt{a^2D(X)}=a\sqrt{D(X)}=a\sigma(X)$







**P.S. 区分标准差（均方差） 与 均方误差**

$MSE$=	mean squared error 

均方误差，它是各个预测数据偏离真实值的差值的平方和的算数平均， 即误差平方和的平均数:

$MSE=\dfrac{1}{n}\sum_{i=1}^n(Y_i-\hat Y_i)^2$

$n$	=	number of data points

$Y_i$	=	observed values

$\hat Y_i$	=	predicted values

## Population variance and sample variance

Real-world observations such as the measurements of yesterday's rain throughout the day typically cannot be complete sets of all possible observations. As such, the variance calculated from the finite set(有限子集) will in general not match the variance that would have been calculated from the full population of possible observations. 
来自样本的方差与总体方差不一样

This means that we can estimates the mean and variance from a limited set(有限子集) of observations by using an estimator equation. The estimator is a function of the sample of n observations drawn from the whole population of potential observations, without observational bias. In this example that sample would be the set of actual measurements of yesterday's rainfall from available rain gauges within the geography of interest.

The simplest estimators for population mean and population variance are simply the mean and variance of the sample
总体均值和总体方差的最简单估计量就是样本的均值和方差

the sample mean and (uncorrected) sample variance – these are consistent estimators (they converge to the correct value as the number of samples increases), but can be improved. 

样本均值和（未校正的）样本方差——这些是一致的估计量，随着样本数量的增加，它们会收敛到正确的值。

**样本均值Sample mean**

$\bar X =\dfrac{1}{n}\sum_{i=1}^{n}X_i$ 意义清楚

**样本方差sample variance**

$S^2=\dfrac{1}{n-1}\sum_{i=1}^{n}(X_i-\bar X)^2$

**样本标准差 sample standard deviation**

$S=\sqrt{S^2}=\sqrt{\dfrac{1}{n-1}\sum_{i=1}^{n}(X_i-\bar X)^2}$

中心极限定理：

$\lim_{n \rightarrow \infty}\bar X \rightarrow \mu$

$\lim_{n \rightarrow \infty}S^2 \rightarrow \sigma^2$


Estimating the population variance by taking the sample's variance is close to optimal in general, but can be improved in two ways. 

Most simply, the sample variance is computed as an average of squared deviations about the (sample) mean, by dividing by n. 

$S^2=\dfrac{1}{n}\sum_{i=1}^{n}(X_i-\bar X)^2$

However, using other values instead of n improves the estimator. Four common values for the denominator are n, n − 1, n + 1, and n − 1.5

1. n is the simplest (population variance of the sample)
2. n − 1 eliminates bias
3. n + 1 minimizes mean squared error(MSE) for the normal distribution
4. n − 1.5 mostly eliminates bias in unbiased estimation of standard deviation for the normal distribution.


Firstly, if the true population mean is unknown($\mu$ is unknown), then the sample variance (which uses the sample mean $\bar X$ in place of the true mean $\mu$) is a biased estimator: it underestimates the variance by a factor of (n − 1) / n. However, we can correct it by this factor (dividing by n − 1 instead of n) is called **Bessel's correction**

The resulting estimator is unbiased, and is called the (corrected) sample variance or unbiased sample variance.（修正样本方差/无偏样本方差）

For example, when n = 1 the variance of a single observation about the sample mean (itself) is obviously zero regardless of the population variance. If the mean is determined in some other way than from the same samples used to estimate the variance, then there is no bias and the variance can safely be estimated as that of the samples about the (independently known) mean.





### Population variance 总体方差

In general, the population variance of a finite population of size $N$ with values $x_i$ is given by

\begin{aligned}
\sigma^2 & =\frac{1}{N} \sum_{i=1}^N\left(x_i-\mu\right)^2=\frac{1}{N} \sum_{i=1}^N\left(x_i^2-2 \mu x_i+\mu^2\right) \\
& =\left(\frac{1}{N} \sum_{i=1}^N x_i^2\right)-2 \mu\left(\frac{1}{N} \sum_{i=1}^N x_i\right)+\mu^2 \\
& = \left(\frac{1}{N} \sum_{i=1}^N x_i^2\right)-2\mu^2 + \mu^2\\
& =\left(\frac{1}{N} \sum_{i=1}^N x_i^2\right)-\mu^2
\end{aligned}


where the population mean is $\mu=\dfrac{1}{N} \sum_{i=1}^N x_i$

The population variance can also be computed using

$\sigma^2=\frac{1}{N^2} \sum_{i<j}\left(x_i-x_j\right)^2=\frac{1}{2 N^2} \sum_{i, j=1}^N\left(x_i-x_j\right)^2$

**Prove**

based on former formula:


$\sigma^2= \frac{1}{N} \sum_{i=1}^N (x_i^2)-\mu^2 \Rightarrow \frac{1}{N} \sum_{i=1}^N (x_i^2)=\sigma^2+\mu^2$


\begin{aligned}
& \frac{1}{2 N^2} \sum_{i, j=1}^N\left(x_i-x_j\right)^2 \\
= & \frac{1}{2 N^2} \sum_{i, j=1}^N\left(x_i^2-2 x_i x_j+x_j^2\right) \\
= & \frac{1}{2 N} \sum_{j=1}^N\left(\frac{1}{N} \sum_{i=1}^N x_i^2\right)-\left(\frac{1}{N} \sum_{i=1}^N x_i\right)\left(\frac{1}{N} \sum_{j=1}^N x_j\right)+\frac{1}{2 N} \sum_{i=1}^N\left(\frac{1}{N} \sum_{j=1}^N x_j^2\right) \\
= & \frac{1}{2}\left(\sigma^2+\mu^2\right)-\mu^2+\frac{1}{2}\left(\sigma^2+\mu^2\right) \\
= & \sigma^2
\end{aligned}

P.S. The population variance matches the variance of the generating probability distribution.(总体方差=分布的方差) In this sense, the concept of population can be extended to continuous random variables with infinite populations.(population 也可以推广到连续型随机变量)


### Sample variance 样本方差

#### Biased sample variance有偏样本方差

In many practical situations, the true variance of a population is not known and must be computed somehow（真实方差是未知的）. When dealing with extremely large populations, it is not possible to count every object in the population, so the computation must be performed on a sample of the population.（当population过于庞大，不可能根据方差定义求出真实值，取而代之的是根据一部分样本来求方差） Sample variance can also be applied to the estimation of the variance of a continuous distribution from a sample of that distribution.

We take a sample with replacement of $n$ values $Y_1, ..., Y_n$ from the population, where $n < N$, and estimate the variance on the basis of this sample. Directly taking the variance of the sample data gives the **average of the squared deviations:**

$\tilde{S}_Y^2=\frac{1}{n} \sum_{i=1}^n\left(Y_i-\bar{Y}\right)^2=\left(\frac{1}{n} \sum_{i=1}^n Y_i^2\right)-2\bar{Y}^2+\bar{Y}^2=\frac{1}{n^2} \sum_{i, j: i<j}\left(Y_i-Y_j\right)^2 $

where $\bar Y$ denotes the sample mean:

$\bar{Y}=\frac{1}{n} \sum_{i=1}^n Y_i$

Since the $Y_i$ are selected randomly, both $\bar{Y}$ and $\tilde{S}_Y^2$ are random variables. Their expected values can be evaluated by averaging over the ensemble of all possible samples $\left\{Y_i\right\}$ of size $n$ from the population. 

For $\tilde{S}_Y^2$ this gives:

\begin{aligned}
\mathrm{E}\left[\tilde{S}_Y^2\right] & =\mathrm{E}\left[\frac{1}{n} \sum_{i=1}^n\left(Y_i-\frac{1}{n} \sum_{j=1}^n Y_j\right)^2\right] \\
& =\frac{1}{n} \sum_{i=1}^n \mathrm{E}\left[Y_i^2-\frac{2}{n} Y_i \sum_{j=1}^n Y_j+\frac{1}{n^2} \sum_{j=1}^n Y_j \sum_{k=1}^n Y_k\right] \\
& =\frac{1}{n} \sum_{i=1}^n\left(\frac{n-2}{n} \mathrm{E}\left[Y_i^2\right]-\frac{2}{n} \sum_{j \neq i} \mathrm{E}\left[Y_i Y_j\right]+\frac{1}{n^2} \sum_{j=1}^n \sum_{k \neq j}^n \mathrm{E}\left[Y_j Y_k\right]+\frac{1}{n^2} \sum_{j=1}^n \mathrm{E}\left[Y_j^2\right]\right) \\
& =\frac{1}{n} \sum_{i=1}^n\left[\frac{n-2}{n}\left(\sigma^2+\mu^2\right)-\frac{2}{n}(n-1) \mu^2+\frac{1}{n^2} n(n-1) \mu^2+\frac{1}{n}\left(\sigma^2+\mu^2\right)\right] \\
& =\frac{n-1}{n} \sigma^2 .
\end{aligned}


Hence $\tilde{S}_Y^2$ gives an estimate of the population variance that is biased by a factor of $\dfrac{n-1}{n}$. For this reason, $\tilde{S}_Y^2$ is referred to as the biased sample variance.



#### Unbiased sample variance无偏样本方差
#### (How to explain $n-1$ in the denominator)



Correcting for this bias yields the unbiased sample variance, denoted $S^2$

$S^2=\dfrac{n}{n-1} \tilde{S}_Y^2=\dfrac{n}{n-1}\left[\dfrac{1}{n} \sum_{i=1}^n\left(Y_i-\bar{Y}\right)^2\right]=\dfrac{1}{n-1} \sum_{i=1}^n\left(Y_i-\bar{Y}\right)^2$


The use of the term $n − 1$ is called **Bessel's correction**, and it is also used in **sample covariance** and the **sample standard deviation (the square root of variance)**. The square root is a concave function and thus introduces negative bias (by Jensen's inequality), which depends on the distribution, and thus the corrected sample standard deviation is biased (using Bessel's correction) . The unbiased estimation of standard deviation is a technically involved problem, though for the normal distribution using the term $n − 1.5$ yields an almost unbiased estimator.





### Variance of mean

To calculate the variance of the mean, you first need to calculate the variance of the population or sample from which the mean is calculated. The variance is a measure of how spread out the data is from the average.

Once you have the variance of the population or sample, you can use the following formula to calculate the variance of the mean:

Variance of the mean = (variance of population or sample) / (sample size)

The sample size is the number of observations used to calculate the mean. If the entire population is used to calculate the mean, then the sample size would be equal to the size of the population.
